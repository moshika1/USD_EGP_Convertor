```python
import requests 
from bs4 import BeautifulSoup 
```

```python
url = 'https://www.exchangerates.org.uk/Dollars-to-Egyptian-Pounds-currency-conversion-page.html'
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text , 'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
convertor = soup.find('span', attrs={'id': 'shd2b;'})
convertor
```

```python
convertor = float(convertor)
convertor
```

```python
usd_amount = int(input('enter usd amount'))
usd_egp_convertor = usd_amount * convertor 
print(f'{usd_amount} Dollars equal {usd_egp_convertor} Egyptian Pounds')
```
